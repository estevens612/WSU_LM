```python
import pandas as pd # data processing
```

```python
# Load faculty dataset
faculty_df = pd.read_csv("https://github.com/estevens612/WSU_LM/blob/main/academic_faculty_data.csv?raw=True")
faculty_df['Areas of Research Interest'].fillna('', inplace=True)
faculty_df['Areas of Teaching Interest'].fillna('', inplace=True)
faculty_df['Other Interests'].fillna('', inplace=True)
faculty_df['Grants'].fillna('', inplace=True)
faculty_df['Academic Interests and Expertise'].fillna('', inplace=True)
faculty_df['Awards and Honors'].fillna('', inplace=True)
faculty_df.head(2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# TF-IDF vectorization
tfidf_vectorizer = TfidfVectorizer()
faculty_df['combined_text'] = faculty_df['Areas of Research Interest'] + ' ' + faculty_df['Areas of Teaching Interest'] + ' ' + faculty_df['Other Interests'] + ' ' + faculty_df['Grants'] + ' ' + faculty_df['Academic Interests and Expertise'] + ' ' + faculty_df['Awards and Honors']

tfidf_matrix = tfidf_vectorizer.fit_transform(faculty_df['combined_text'])

# Compute similarity matrix
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Example: Recommend top 3 faculty based on user's interests
user_interests = "machine learning"
user_vector = tfidf_vectorizer.transform([user_interests])
similarity_scores = cosine_similarity(user_vector, tfidf_matrix)[0]
sorted_indices = similarity_scores.argsort()[::-1]

# Recommend top 5 similar faculty
top_faculty_indices = sorted_indices[:5]
recommended_faculty = faculty_df.iloc[top_faculty_indices]['Name']
print(recommended_faculty)
```
